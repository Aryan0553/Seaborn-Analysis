```python
#kernal density (kde) a method for visualizing the distribution of observations in a dataset, analogous to a histogram
#rug parameter (rug) shows which path the kde is following in graph
```

```python
data=sns.load_dataset("penguins")
b=[170,180,190,200,210,220,230,240,250]
sns.displot(data["flipper_length_mm"],bins=b,kde=True,rug=True,color='r')
plt.show()
```

```python
data=sns.load_dataset("penguins")
b=[170,180,190,200,210,220,230,240,250]
sns.displot(data["flipper_length_mm"],log_scale=True,kde=True,rug=True,color='r')
plt.show()
```

```python
data=sns.load_dataset("penguins")
data
```

```python
sns.barplot(x=data.island,y=data.bill_length_mm,hue="sex",data=data)
plt.show()
```

```python
sns.set(style="darkgrid")
ord=["Dream","Torgersen","Biscoe"]
sns.barplot(x=data.island,y=data.bill_length_mm,hue="sex",data=data,order=ord,hue_order=["Female","Male"],errorbar=('ci', 100),palette="Oranges",
            saturation=1,errcolor="m",errwidth=5,capsize=0.9)
plt.show()
```
